```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("C:/Users/User/Downloads/SME_CA_SOLUCAO.csv", sep='\t', dtype='str')
```

```python
n1 = []
for c in data['Categoria'].value_counts().axes[0].tolist():
    if '.' in c:
        i = c.find('.')
        n1.append(c)
n1 = np.array(n1)
categorias = np.unique(n1)
print(categorias)
```

```python
data.head()
```

```python
categorias_chatbot = []
for i in categorias:
    if ("Escola 3" in i or "Software" in i or "Sistemas" in i or "Software" in i) and ('Reset' not in i and 'reset' not in i):
        categorias_chatbot.append(i)
categorias_chatbot = np.array(categorias_chatbot)
categorias_chatbot = np.setxor1d(categorias, categorias_chatbot)
categorias_chatbot
```

```python
for categoria in categorias_chatbot:
    data = data.drop(data.loc[data['Categoria'] == categoria].index)
```

```python
data['Categoria'].value_counts().axes[0]
```

```python
metodos_remover = np.array(data['Metodo de Resolucao'].value_counts().axes[0].tolist())
metodos_remover = np.setxor1d(metodos_remover, np.array(['Suporte Telefônico', 'Suporte por redes sociais', 'Suporte por e-mail']))
metodos_remover
```

```python
for metodo in metodos_remover:
    data = data.drop(data.loc[data['Metodo de Resolucao'] == metodo].index)
```

```python
def busca_categoria_palavrachave(keyword):
    for i in data['Categoria'].value_counts().axes[0].tolist():
        if str.lower(keyword) in str.lower(i):
            print(i)
busca_categoria_palavrachave('')
```

```python
categorias_catalogo = ['Sistemas Departamentais.Ergon', 'Sistemas Departamentais.Fincon', 'Sistemas Departamentais.Sisbens', 'Sistemas Departamentais.Desesc', 'Sistemas Departamentais.Sigma', 'Sistemas Departamentais.Fenix', 'Sistemas Corporativos.Processo.Rio']
```

```python
l = 0
for categoria in categorias_catalogo:
    l+=len(data.loc[data['Categoria'] == categoria])
print(l)
len(data)
```

```python
data_softwares = []
for categoria in categorias_catalogo:
    data_softwares.append(data.loc[data['Categoria'] == categoria])
data_softwares = np.vstack(data_softwares)
data_softwares = pd.DataFrame(data=data_softwares, columns=data.columns)
data_softwares.head()
```

```python
data_softwares.loc[data_softwares['Metodo de Resolucao'] == 'Acesso servidor']
```

```python
top20 = pd.DataFrame(data_softwares[['Descricao', 'Solução']].value_counts(), columns=['solucao', 'count']).sort_values('count', ascending=False)[0:20]
```

```python
def categorias_primeiras5(categorias):
    for categoria in categorias:
        print('---------------------------------\n'+categoria+":\n")
        for reg in data_softwares['Solução'].value_counts().axes[0].tolist()[0:5]:
            print(reg+";\n")
categorias_primeiras5(categorias_catalogo)
```

```python
def busca_descricao_porSolucao(solucao, categorias):
        for categoria in categorias:
            print('---------------------------------\n'+categoria+":\n")
            for reg in data_softwares.loc[data['Solução']==solucao]['Descricao']:
                print(reg+";\n")
busca_descricao_porSolucao("sistema instalado.", categorias_catalogo)
```
